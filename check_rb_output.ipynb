```python
# REQUIREMENTS
from os import listdir

# LISTDIR_NH FUNCTION
def listdir_nh(path):
    '''
    Input: Path of directory (string)
    Output: Non-hidden files within given directory (list of strings)
    '''
    files = listdir(path)
    parsed = [file for file in files if not file.startswith(".")]
    return parsed
```

## 2. Outcome Checker

```python
def checkFailures(script_path):
    # READ FILE AS STRING
    with open (script_path, "r") as this_file:
        data = this_file.read()
        
    # CHECK IF PARTICULAR STRINGS IN IT
    if "FAILURE" in data:
        return("Failure")
    if "Error:" in data:
        return("Error")
    
    return("")
```

```python
outcomes = [checkFailures(x) for x in file_paths]
```

```python
outcome_table = pd.DataFrame([names[:len(outcomes)], tasklines[0:len(outcomes)], outcomes], ["dataset", "script", "outcome"]).T
```

```python
outcome_table_sorted = outcome_table.sort_values("outcomes", ascending=False)
```

# 3. Total Processing Time Checker

```python
def checkProcessingTime(script_path):
    # READ FILE AS STRING
    with open (script_path, "r") as this_file:
        data = this_file.read()
    data = data.split("\n")
        
    # CHECK TIME
    try:
        for i in range(len(data)):
            if "Total processing time" in data[i]:
                final_line = i
        return(data[final_line].split(": ")[1])
    except:
        return("")
```

```python
times = [checkProcessingTime(x) for x in file_paths]
```

```python
# PREVIEW SOME OF THE TIMES
times[0:30:2]
```

```python
outcome_time_table = pd.DataFrame([names[:len(outcomes)], tasklines[:len(outcomes)], times, outcomes], ["dataset", "script", "hh:mm:ss", "outcome"]).T
```
